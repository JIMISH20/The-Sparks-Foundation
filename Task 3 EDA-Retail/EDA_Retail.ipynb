**Using Seaborn Libraries : Ploting informative statistical Graphics**

```python
sns.pairplot(df,height=2.0 , diag_kind = "kde");
```

```python
plt.figure(figsize=(6, 8))
splot = sns.countplot(x=df['Ship Mode'],order=df['Ship Mode'].value_counts().index,)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

```python
df['Ship Mode'].value_counts()
```

```python
plt.figure(figsize=(6, 8))
splot = sns.countplot(x=df['Segment'])
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

```python
df['Segment'].value_counts()
```

```python
plt.figure(figsize=(6, 8))
splot = sns.countplot(x=df['Category'], order=df['Category'].value_counts().index)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

```python
df['Category'].value_counts()
```

```python
sns.countplot(x=df['Sub-Category'] ,  order=df['Sub-Category'].value_counts().index)
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(10,10));
sns.countplot(x=df['State'] ,  order=df['State'].value_counts().index)
plt.xticks(rotation=90);
```

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Region',data=df,palette='Set1',order=df['Region'].value_counts().index)
```

```python
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='Set2')
plt.xlabel('Region',fontsize=15)
plt.show()
```

**Final Observations :-**
